```python
import pandas as pd
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
from datetime import datetime
from meteostat import Point, Daily
```

```python
cities_coordinates = {
    "Berlin": [52.5200, 13.4050],
    "Hamburg": [53.5511, 9.9937],
    "Munich": [48.1351, 11.5820],
    "Frankfurt": [50.1109, 8.6821],
    "Cologne": [50.9375, 6.9603],
    "Stuttgart": [48.7758, 9.1829],
    "Dresden": [51.0504, 13.7373],
    "Leipzig": [51.3397, 12.3731],
    "Rostock": [54.0924, 12.0991],
    "Freiburg": [47.9990, 7.8421]
}
```

```python
wormplot_weather = pd.DataFrame()
```

```python
start = datetime(2009, 1, 1)
end = datetime(2023, 12, 31)
weather_data = {}
```

```python
for city in cities_coordinates:
    
   

   
    weather_data[city] = Daily(Point(cities_coordinates[city][0],cities_coordinates[city][1]), start, end)
    weather_data[city] = weather_data[city].fetch()
    weather_data[city].fillna(weather_data[city].mean(), inplace=True)
    
    weather_data[city]["city_name"] = city
    weather_data[city].reset_index(inplace=True)
    weather_data[city]['time'] = pd.to_datetime(weather_data[city]['time']) 
    weather_data[city]['day'] = weather_data[city]['time'].dt.day 
    weather_data[city]['time'] = weather_data[city]['time'].dt.to_period('M')
    weather_data[city]['time_int'] = weather_data[city]['time'].rank(method='dense').astype(int) 
    wormplot_weather = pd.concat([wormplot_weather, weather_data[city]], axis=0)
    
```

```python
wormplot_weather = pd.DataFrame()
```

```python
columns_to_scale = ['tavg', 'tmin','tmax','prcp','snow','wdir','wspd','wpgt','pres','tsun']
```

```python
start = datetime(2009, 1, 1)
end = datetime(2023, 12, 31)
weather_data = {}
```

```python
for city in cities_coordinates:
    
   

   
    weather_data[city] = Daily(Point(cities_coordinates[city][0],cities_coordinates[city][1]), start, end)
    weather_data[city] = weather_data[city].fetch()
    weather_data[city].fillna(weather_data[city].mean(), inplace=True)
    
    # Initialize the MinMaxScaler
    scaler = MinMaxScaler()

    # Fit and transform the selected columns
    weather_data[city][columns_to_scale] = scaler.fit_transform(weather_data[city][columns_to_scale])*10

    weather_data[city]["city_name"] = city
    weather_data[city].reset_index(inplace=True)
    weather_data[city]['time'] = pd.to_datetime(weather_data[city]['time']) 
    weather_data[city]['day'] = weather_data[city]['time'].dt.day 
    weather_data[city]['time'] = weather_data[city]['time'].dt.to_period('M')
    weather_data[city]['time_int'] = weather_data[city]['time'].rank(method='dense').astype(int) 
    wormplot_weather = pd.concat([wormplot_weather, weather_data[city]], axis=0)
    
```

```python
wormplot_weather.to_csv('wormplot_weather_scaled-10.csv', index = False)
```
